## Two plots in one figure

```python
figure, axes_set = plt.subplots(2)
```

```python
axes_set
```

```python
x = np.linspace(1, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='red')
axes_set[1].plot(x, y2, color='green')
```

```python
x = np.linspace(1, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='red')
axes_set[0].annotate('breakout:', xy=(0,0), xytext=(1,2000),
                    arrowprops={'facecolor': 'black'})

axes_set[1].plot(x, y2, color='green')
```

```python
x = np.linspace(1, 100, 1000)
y1 = x ** 2
y2 = x ** 3

figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='red', label='main 1 line')
axes_set[0].legend(title='Legend', loc='lower right', shadow=True)
axes_set[0].annotate('breakout:', xy=(0,0), xytext=(1,2000),
                    arrowprops={'facecolor': 'black'})

axes_set[1].plot(x, y2, color='green', label='main 2 line')
axes_set[1].legend()
```

## Exercise 7

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure, axes_set = plt.subplots(4)

ymax = max(y)
xpos = np.where(y == ymax)
xmax = x[xpos]

axes_set[0].plot(x, y, color='blue')
axes_set[0].set_title('y against x')

axes_set[1].plot(x, -y, color='orange')
axes_set[1].set_title('-y against x')

axes_set[2].plot(x, y ** 2, color='green')
axes_set[2].set_title('y squared against x')

axes_set[3].plot(x, -(y ** 2), color='red')
axes_set[3].set_title('-y squared against x')

plt.tight_layout()
```

## Exercise 8

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

figure, axes_set = plt.subplots(4)

axes_set[0].plot(x, y, color='blue')
axes_set[0].set_title('y against x')

max_y = max(y)
min_y = min(y)
axes_set[0].annotate(f'Max: {max_y:.2f}', xy=(x[np.argmax(y)], max_y), xytext=(x[np.argmax(y)], max_y - 0.5),
                        arrowprops=dict(arrowstyle='->'))


axes_set[1].plot(x, -y, color='orange')
axes_set[1].set_title('-y against x')

axes_set[2].plot(x, y ** 2, color='green')
axes_set[2].set_title('y squared against x')

axes_set[3].plot(x, -(y ** 2), color='red')
axes_set[3].set_title('-y squared against x')

plt.tight_layout()
```
