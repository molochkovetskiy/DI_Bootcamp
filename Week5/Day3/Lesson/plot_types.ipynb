```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Line plot - ax.plot

```python
figure, axes = plt.subplots()
axes.plot([1,2,3,4], [2,4,5,7])
```

## Scatter plot - individual data points

```python
x = np.random.rand(100)
y = np.random.rand(100)

figure, axes = plt.subplots()
axes.scatter(x,y)
```

## Histogram - putting the data into bins based on value

```python
# data = np.random.rand(1000)
data = [1,1,1,5,5,5,5,5,5,5,5,9,9,9,9]

figure, axes = plt.subplots()
axes.hist(data, bins=3)
```

## Styling graphs

```python
figure, axes = plt.subplots()
axes.set(title='Line plot', xlabel='X', ylabel="Y")
axes.plot([1,2,3,4], [2,4,5,7])
```

```python
figure, axes = plt.subplots()
axes.set(title='Line plot', xlabel='X', ylabel="Y")
axes.plot([1,2,3,4], [2,4,5,7], color='green', linestyle='--')
```

```python
figure, axes = plt.subplots()
axes.set(title='Line plot', xlabel='X', ylabel="Y")
# marker
axes.plot([1,2,3,4], [2,4,5,7], color='green', linestyle='--', marker='o', mfc='red', mec='blue')
```

```python
figure, axes = plt.subplots()
axes.set(title='Line plot', xlabel='X', ylabel="Y")

axes.plot([1,2,3,4], [2,4,5,7], color='green', linestyle='--', marker='o', mfc='red', mec='blue', label='Yossi')
axes.plot([6,3,4,10], [5,12,4,9], color='red', linestyle='--', marker='*', mfc='blue', mec='green', label='Ziv')

axes.legend()
```

## Saving plots

```python
figure, axes = plt.subplots()
axes.set(title='Line plot', xlabel='X', ylabel="Y")

axes.plot([1,2,3,4], [2,4,5,7], color='green', linestyle='--', marker='o', mfc='red', mec='blue', label='Yossi')
axes.plot([6,3,4,10], [5,12,4,9], color='red', linestyle='--', marker='*', mfc='blue', mec='green', label='Ziv')

figure.savefig('my_plot.png')
```

## Exercise 1

```python
x = np.linspace(-10, 10, 100)
y = x ** 2

figure, axes = plt.subplots()

axes.plot(x, y)
```

## Exercise 2-3

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5

figure, axes = plt.subplots()

axes.plot(x, y1)
axes.scatter(x, y2)

figure.savefig('basic_plots.png')
```

## Exercise 4

```python
x = np.linspace(0, 10, 20)
y1 = x ** 2
y2 = x ** 1.5

figure, axes = plt.subplots()
axes.set(title='My Styled Plot', xlabel='X', ylabel="Y")

axes.plot(x, y1, color='red', marker='o', linestyle='dashed', label='y = x^2')
axes.plot(x, y2, color='blue', marker='s', linestyle='dotted', label='y = x^1.5')

axes.legend()
figure.savefig('my_styled_plot.png')
```
