# Task 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 20, 100)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title('Plot of x vs y (y=x^2)')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0, 20)
plt.ylim(0, 400)

plt.savefig('x_vs_y_plot.png')
```

# Task 2

```python
chipo = pd.read_csv('chipo.txt', sep='\t')
```

```python
chipo
```

```python
chipo['item_price'] = chipo['item_price'].apply(lambda row: float(row[1:]))
```

```python
item_price = chipo['item_price']
quantity = chipo['quantity']

figure, axes = plt.subplots()

axes.set(title='Relationship between Item Price and Quantity Ordered', xlabel='Item Price ($)',
         ylabel='Quantity Ordered')
axes.scatter(item_price, quantity)
axes.grid(True)

figure.savefig('task2.png')
```

# Task 3

```python
item_quantities = chipo.groupby('item_name')['quantity'].sum()
item_quantities
```

```python
top_items = item_quantities.nlargest(5)
top_items
```

```python
figure, axes = plt.subplots()

axes.set(title='Top 5 Items by Quantity Ordered', ylabel='Total Quantity Ordered')
axes.bar(top_items.index, top_items.values)
axes.tick_params(axis='x', labelrotation=45)

figure.tight_layout()

figure.savefig('task31.png')
```
