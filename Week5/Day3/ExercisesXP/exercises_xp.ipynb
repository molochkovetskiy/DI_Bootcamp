## Exercise 7

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
figure, axes = plt.subplots()
axes.set(title='y = sin(x)')

axes.plot(x, y1, color='tab:blue')
```

```python
figure, axes = plt.subplots()
axes.set(title='y = sin(x)')
axes.scatter(x, y2, color='green', marker='d')
```

## Exercise 8

```python
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='tab:blue')
axes_set[1].scatter(x, y2, color='green', marker='d')

plt.tight_layout()
```

## Exercise 9

```python
figure, axes_set = plt.subplots(2)

axes_set[0].plot(x, y1, color='tab:blue', label='y = sin(x)')
axes_set[0].legend(loc='upper left')

axes_set[1].scatter(x, y2, color='green', marker='d', label='y = cos(x)')
axes_set[1].legend(loc='center right')

plt.tight_layout()
```
