```python
data1 = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': [1, 2, 3, 4],
    'D': ['D0', 'D1', 'D2', 'D3']
}, index=[0, 1, 2, 3])

data2 = pd.DataFrame({
    'A': [5, 6, 7, 8],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': [5, 6, 7, 8],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[4, 5, 6, 7])

data3 = pd.DataFrame({
    'A': [9, 10, 11, 12],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': [9, 10, 11, 12],
    'D': ['D4', 'D5', 'D6', 'D7']
}, index=[0, 1, 2, 3])
```
