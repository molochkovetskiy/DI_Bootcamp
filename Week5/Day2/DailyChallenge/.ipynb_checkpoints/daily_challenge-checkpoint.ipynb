### 11.

```python
starbucks['Caffeine (mg)'] = pd.to_numeric(starbucks['Caffeine (mg)'], errors='coerce')
```

```python
starbucks['caffeine_to_calories'] = starbucks['Caffeine (mg)'] / starbucks['Calories']
```

```python
starbucks
```

### 12.

```python
average_caffeine_per_category = starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean().to_frame()
average_caffeine_per_category
```

```python
average_caffeine_per_category.sort_values(by='caffeine_to_calories', ascending=False).head(1)
```

### 13.

```python
import matplotlib.pyplot as plt
```

```python
average_caffeine_per_category = starbucks.groupby('Beverage_category')['caffeine_to_calories'].mean()
```

```python
plt.figure(figsize=(10, 6))
average_caffeine_per_category.sort_values(ascending=True).plot(kind='barh', color='c')
plt.xlabel('Average Caffeine-to-Calories Ratio')
plt.ylabel('Drink Category')
plt.title('Average Caffeine-to-Calories Ratio for Each Drink Category')
plt.tight_layout()

plt.show()
```

### 14.

```python
average_sugar_per_category = starbucks.groupby('Beverage_category')[' Sugars (g)'].mean()
average_caffeine_per_category = starbucks.groupby('Beverage_category')['Caffeine (mg)'].mean()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(average_sugar_per_category, average_caffeine_per_category, color='purple')
plt.xlabel('Average Sugar Content (g)')
plt.ylabel('Average Caffeine Content (mg)')
plt.title('Relationship Between Sugar and Caffeine Across Different Drink Categories')
plt.tight_layout()

plt.show()
```
